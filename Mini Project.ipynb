## Project on IPL Data Set

```python
import pandas as pd
df=pd.read_csv('Match.csv')
print(df)
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
## Matches based on the year wise
a=df.groupby('Season_Year')['Country_id'].count()
print(a)
```

```python
## Bar graph no.of matches won by each team in the year 2011
b=df[df['Season_Year']==2011].groupby('match_winner')['Country_id'].count()
q=b.index.tolist()
w=b.values.tolist()
plt.bar(q,w,color='blue',width=0.5)
```

```python
## ManOfMatch count of each player in Hyderabad Stadium
b=df[df['City_Name']=='Hyderabad'].groupby('ManOfMach')['Country_id'].count()
b
```
